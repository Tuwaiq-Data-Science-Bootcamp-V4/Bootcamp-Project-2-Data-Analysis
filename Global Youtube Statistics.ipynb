```python
# drop unnecessary column
dfClean=dfClean.drop(columns=['Latitude','Longitude','Gross tertiary education enrollment (%)',
                              'Abbreviation','Urban_population','subscribers_for_last_30_days',
                              'country_rank','Population','Unemployment rate'])
```

```python
# check for null value
dfClean.isnull().sum()
```

Dealing with missing values

```python
dfClean.drop(index=dfClean[(dfClean.category.isnull())&(dfClean.channel_type.isnull())].index,inplace=True)
dfClean
```

```python
dfClean.category.unique()
```

```python
dfClean.channel_type.unique()
```

```python
# Now for filling the missing value in category columns we will make a dictionary
category_dict = {'Music':'Music','Games':'Gaming','Entertainment':'Entertainment','Education':'Education','People':'People & Blogs','Sports':'Sports',
                'Film':'Film & Animation','News':'News & Politics','Comedy':'Comedy','Howto':'Howto & Style','Nonprofit':'Nonprofits & Activism',
                'Autos':'Autos & Vehicles','Tech':'Science & Technology','Animals':'Pets & Animals'}
inx=dfClean[dfClean.category.isnull()].index
dfClean['category'][inx] = dfClean.loc[inx,'channel_type'].apply(lambda x: category_dict[x])

#filling the missing value in video_views_for_the_last_30_days using mean
dfClean['video_views_for_the_last_30_days'].fillna(dfClean['video_views_for_the_last_30_days'].mean(), inplace = True)
```

```python
#drop unnecessary column
dfClean=dfClean.drop(columns=['channel_type'])

# Converting NULL values to "unknown"
dfClean = dfClean.fillna('unKnown')

#drop all null values in created_year
dfClean = dfClean.drop(dfClean[dfClean['created_year'] == 'unKnown'].index)
```

```python
#check for null value after cleaning
dfClean.isnull().sum()
```

```python
#shape after cleaning data
dfClean.shape
```

```python
#conver created_year type to integer
dfClean['created_year'] = dfClean['created_year'].astype(np.int64)
dfClean['created_date'] = dfClean['created_year'].astype(np.int64)
```

```python
# summary of our clean dataset.
dfClean.info()
```

# adding new column and some addational analysis

```python
#replace shortcuts in the month column to be meaningful
dfClean.created_month=dfClean.created_month.replace(to_replace=['Jan', 'Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov', 'Dec'],
             value=['January','February','March','April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
```

```python
dfClean['subscribers_per_video'] = np.divide(dfClean['subscribers'], dfClean['uploads'])
```

```python
          
month_mapping = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 'June': 6,
    'July': 7, 'August': 8, 'September': 9, 'October': 10, 'November': 11, 'December': 12
}
```

```python
current_year = pd.Timestamp.now().year
current_month = pd.Timestamp.now().month
```

```python
dfClean['channel_age_months'] = (current_year - dfClean['created_year']) * 12 + (current_month - dfClean['created_month'].map(month_mapping))
```

```python
age_bins = [0, 6, 12, 24, float('inf')]
age_labels = ['Emerging', 'Established', 'Influential', 'Iconic']

dfClean['channel_age_category'] = pd.cut(dfClean['channel_age_months'], bins=age_bins, labels=age_labels)
```

```python
Country_dict = {'India':'Asia','United States':'North America','unKnown':'unKnown', 'Japan':'Asia', 
                 'Russia':'Asia','South Korea':'Asia','United Kingdom':'Europe','Canada':'North America', 
                 'Brazil':'South America','Argentina':'South America','Chile':'South America', 
                 'Cuba':'North America', 'El Salvador':'North America', 'Pakistan':'Asia', 'Philippines':'Asia',
                 'Thailand':'Asia', 'Colombia':'South America', 'Barbados':'North America', 
                 'Mexico':'North America','United Arab Emirates':'Asia', 'Spain':'Europe', 'Saudi Arabia':'Asia', 
                 'Indonesia':'Asia','Turkey':'Asia', 'Venezuela':'South America', 'Kuwait':'Asia', 
                 'Jordan':'Asia', 'Netherlands':'Europe','Singapore':'Asia','Australia':'Australia', 
                 'Italy':'Europe','Germany':'Europe', 'France':'Europe', 'Sweden':'Europe',
                 'Afghanistan':'Asia', 'Ukraine':'Europe', 'Latvia':'Europe', 'Switzerland':'Europe', 
                 'Vietnam':'Asia','Malaysia':'Asia', 'China':'Asia', 'Iraq':'Asia', 'Egypt':'Africa', 
                 'Andorra':'Europe', 'Ecuador':'South America','Morocco':'Africa', 'Peru':'South America', 
                 'Bangladesh':'Asia', 'Finland':'Europe', 'Samoa':'Oceania'}

dfClean['continent'] = dfClean['Country'].map(Country_dict)
```

```python
dfClean['avg earnings yearly'] = df[["lowest_yearly_earnings","highest_yearly_earnings"]].mean(axis=1)
```

```python
# Choose a specific style for your charts.
sns.set(style="darkgrid")
# Apply one color palette from your choice on all charts.
co=['#ff9999','#66b3ff','#99ff99','#FFFACD','#FFB6C1','#C8A2C8','#FAF0E6','#FBEC5D','#CC3333','#997A8D']

p=sns.set_palette(sns.color_palette(co))
```

### What are the top 10 countries with the most number of channels?

```python
channel_country_distribution = dfClean[dfClean['Country']!='unKnown']['Country'].value_counts()
topCountries = channel_country_distribution[:10]

# Plot size
plt.figure(figsize=(10, 6))

# Create a bar plot for the number of channels in each country
sns.barplot(x=topCountries.index, y=topCountries.values)
plt.xticks(rotation=90)
# Title and labels
plt.title('Top 10 Countries by Number of Channels',fontsize = 20)
plt.xlabel('Country',fontsize = 20)
plt.ylabel('Number of Channels',fontsize = 20)
plt.show()
```

Insight: It is clear that United State has most number of channels, it has more than 300 channels

### What is the most common channel type worldwide?

```python
plt.figure(figsize=(10, 6)) 
sns.countplot(x='category', order=df['category'].value_counts().index,data=dfClean,color=p)
plt.xticks(rotation=90)
plt.xlabel('Channel Type',fontsize = 20)
plt.ylabel('Counts',fontsize = 20)
plt.title('Channel Type Counts',fontsize = 20)
plt.show()
```

Insight: Entertaiment category is the most channel type on Youtube however, travel&Event is the least type

### What continent has the most YouTube channels, and what are the most popular categories on each continent?

```python
plt.figure(figsize=(30,15))
sns.histplot(dfClean,x="continent",hue="category", multiple="stack")
plt.xlabel('Continent')
plt.title('Distribution of YouTube channels and thier categories for each continent',fontweight='bold',fontsize = 20)
plt.show()
```

- It is obvious that North America has the most YouTube channels of any region.
- Entertainment and music are the most popular industries throughout most of the continent.
- Africa and Oceania have the fewest YouTube channels.

### What are the most popular categories according to the number of subscribers?

```python
plt.figure(figsize=(35,15))
sns.lineplot(data=dfClean.groupby(['category'])['subscribers'].sum())
plt.title('Top categories according to the number of subscribers',fontweight='bold',fontsize = 20)
plt.xlabel('Category')
plt.show()
```

- The most popular categories on YouTube are entertainment and music.
- People rarely utilize YouTube as a platform for watching shows and movies, as seen by the low number of subscribers.
- Youtube is a platform for documenting things through videos, as evidenced by the fact that People & Blogs is the third most popular category.

### What are the categories for the highest-earning channels?

```python
plt.figure(figsize=(30,15))
sns.lineplot(data=dfClean.groupby(['category'])['avg earnings yearly'].mean())
plt.title('Categories for the highest-earning channels',fontweight='bold',fontsize = 20)
plt.xlabel('Category')
plt.show()
```

-  Even though show is one of the least populous categories, it is one of the most profitable, as are Autos & Vehicles.
- Also, you'd think that entertainment would be one of the highest-paying categories given the number of subscribers that these channels bring to YouTube, but as it turns out, it pays less than a lot of other, less popular categories.
