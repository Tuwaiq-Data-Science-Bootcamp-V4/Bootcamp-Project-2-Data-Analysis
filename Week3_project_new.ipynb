## 4. Data Cleaning & Preprocessing

```python
#Drop Uneeded columns
dropedDf = df.drop(['index', 'Headquarters', 'Competitors'], axis = 1)
dropedDf
```

```python
def clean_ratings(rating_str):
    low, high = rating_str.split('-')
    low = low.replace('$', '').replace('K', '').strip()
    high = high.replace('$', '').replace('K', '').replace('(Glassdoor est.)', '').strip()
    
    return f'{low} - {high}'

dropedDf['Salary Estimate'] = dropedDf['Salary Estimate'].apply(clean_ratings)
dropedDf['Salary Estimate'] = dropedDf['Salary Estimate'].str.replace(r'\(Employer est.\)', '', regex=True)
dropedDf['Salary Estimate']
```

```python
dropedDf = dropedDf[dropedDf != '-1']
dropedDf = dropedDf[dropedDf != -1]
dropedDf
```

```python
dropedDf.isnull().sum()
```

```python
ndf = dropedDf.dropna().reset_index(drop = True)
ndf
```

```python
ndf['Founded'] = ndf['Founded'].astype(int)
ndf
```

```python
#pd.set_option('display.max_rows', None)
ndf['Size'] = ndf['Size'].str.replace(r'(\d+) to (\d+) employees', r'\1 - \2')
ndf['Size'] = ndf['Size'].str.replace(r' employees', '').str.replace(r'(\d+)\+', r'+\1')
ndf
```

```python
ndf[['Start', 'End']] = ndf['Salary Estimate'].str.split('-', expand=True)
ndf['Start'] = pd.to_numeric(ndf['Start'])
ndf['End'] = pd.to_numeric(ndf['End'])
ndf['Salary Estimate'] = (ndf['Start'] + ndf['End']) / 2
ndf.drop(columns=['Start', 'End'],inplace=True)
```

```python
ndf
```

```python
ndf = ndf.assign(**{column: 0 for column in ['Python', 'SQL', 'Alteryx','ML', 'Statistics']})
ndf
```

```python
new_columns = ['Python', 'SQL', 'Alteryx','ML', 'Statistics','Tableau']
for column in new_columns:
    if column.lower() == 'ml':
        ndf[column] = ndf['Job Description'].apply(lambda x: 1 if 'machine learning' in x.lower() else 0)
    else:
        ndf[column] = ndf['Job Description'].apply(lambda x: 1 if column.lower() in x.lower() else 0)
ndf['Total Skills'] = ndf[new_columns].sum(axis=1)
ndf[['Python', 'SQL', 'Alteryx', 'ML', 'Statistics', 'Tableau','Total Skills']]
```

```python
ndf = ndf.drop(['Job Description'], axis = 1)
ndf
```

## 5. Data Visualization

### 1. Salary Density

```python
plt.figure(figsize=(18,8))
sns.kdeplot(ndf['Salary Estimate'], 
            color='#39185c',
            shade=True).set_title('DS Job Salaries',fontdict = { 'fontsize': 18});
plt.ylabel('Density', fontsize=15)
plt.xlabel('Salary',fontsize=15)
plt.show()
```

### 2.Number of data science jobs in each sectore

```python
plt.figure(figsize=(18,8))
# Sectors = ndf['Sector'].value_counts().sort_values(ascending=False);
sns.countplot(y=ndf['Sector'], data=ndf, palette = 'viridis');
plt.title('DS Jobs in Various Sectors', fontsize=18)
plt.ylabel('Sctors', fontsize=15)
plt.xlabel('Number of Jobs',fontsize=15)
plt.show()
```

### 3. Rating of Data Science Jobs 

```python
ndf['Rating_ct'] = ndf['Rating'].apply(lambda x: '1' if x < 2.0
                                      else '2' if x >= 2.0 and x < 3.0
                                      else '3' if x >= 3.0 and x < 4.0
                                      else '4' if x >= 4.0 and x < 5.0
                                      else '5' if x == 5.0 else '' )
plt.figure(figsize=(18,8))
sns.histplot(
    data = ndf,
    x="Size",
    hue = 'Rating_ct',
    multiple="stack",
    palette = 'viridis',
    );
plt.title('Rating of DS Jobs in Various Company Size', fontsize=18)
plt.xlabel('Size', fontsize=15)
plt.ylabel('Number of Jobs',fontsize=15)
plt.show()
```

### 4. Salaries of Data Science Jobs in Sector

```python
sns.catplot(data=ndf, y="Salary Estimate", x="Sector",palette = 'viridis', kind="violin",height=10,aspect=2)
plt.xticks(rotation=90);
plt.title('DS Jobs in Salaries in Various Sector', fontsize=18)
plt.xlabel('Sectors', fontsize=15)
plt.ylabel('Salaries',fontsize=15)
plt.show()
```

### 5. Skills required in Data Science Jobs 

```python
Python = len(ndf.loc[ndf['Python']==1,'Python'])
Sql = len(ndf.loc[ndf["SQL"]==1,"SQL"])
Alteryx = len(ndf.loc[ndf["Alteryx"]==1,"Alteryx"])
Tableau = len(ndf.loc[ndf["Tableau"]==1,"Tableau"])
ML = len(ndf.loc[ndf["ML"]==1,"ML"])
Statistics = len(ndf.loc[ndf["Statistics"]==1,"Statistics"])

skills_r1= [Python,Sql,Alteryx,ML,Statistics,Tableau]

Python0 = len(ndf.loc[ndf['Python']==0,'Python'])
Sql0 = len(ndf.loc[ndf["SQL"]==0,"SQL"])
Alteryx0 = len(ndf.loc[ndf["Alteryx"]==0,"Alteryx"])
Tableau0 = len(ndf.loc[ndf["Tableau"]==0,"Tableau"])
ML0 = len(ndf.loc[ndf["ML"]==0,"ML"])
Statistics0 = len(ndf.loc[ndf["Statistics"]==0,"Statistics"])

skills_r0= [Python0,Sql0,Alteryx0,ML0,Statistics0,Tableau0]
```

```python
x = np.arange(len(skills_r1))
width = 0.35
fig, ax = plt.subplots(figsize=(18, 5))

sns.set_palette(sns.color_palette('viridis'))
rects1 = ax.bar(x - width/2, skills_r1, width, label='Required')
rects2 = ax.bar(x + width/2, skills_r0, width, label='Not-Required')
ax.set_xlabel('Number of Jobs', fontsize=15)
ax.set_ylabel('Skills', fontsize=15)
ax.set_title('DS Skills Rate', fontsize=18)
ax.legend()
ax.set_xticks(x)
ax.set_xticklabels(['Python','SQL','Alteryx','ML','Statistics','Tableau'])
plt.show()
```
